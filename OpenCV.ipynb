```
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import urllib
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')
```

```
img = cv.imread('/content/gdrive/My Drive/openCV.png', -1)
plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```
#Resize
img_resized = cv.resize(img,(500,400))
plt.imshow(img_resized, 'gray')
```

```
#Rotate
(h, w) = img.shape[:2]
# calculate the center of the image
center = (w / 2, h / 2)
 
angle90 = 90
angle180 = 180
angle270 = 270
 
scale = 1.0

M = cv.getRotationMatrix2D(center, angle270, scale)
rotated270 = cv.warpAffine(img, M, (w, h))

plt.imshow(rotated270, 'gray')
```

```
#saving 
cv.imwrite('/content/gdrive/My Drive/Colab Notebooks/ - OpenCV2.png',rotated270)
```
